```python
import re


fname1 = 'tlg0032.tlg006.rouse-eng.xml'

words1 = []
words2 = []
curcit = ''
citpat = "^7\.6"
f = open(fname1)
for l in f:
    m = re.search('subtype="book" n="([0-9]+)">',l)
    if(m):
        curbook = m[1]
    m = re.search('subtype="chapter" n="([0-9]+)">',l)
    if(m):
        curchap = m[1]
        words1.append(curbook + '.' + curchap)
    m = re.search('subtype="section" n="([0-9]+)">',l)
    if(m):
        cursect = m[1]
        curcit = curbook + '.' + curchap + '.' + cursect
    l = re.sub('<[^>]+>',' ',l)
    if(not re.search(citpat,curcit)):
        continue
    for foo in l.split():
        
        words1.append(foo)

f.close()

outf = open('gpt0','w')
for foo in words1:
    print(foo,file=outf)
outf.close()

f = open('tlg0032.tlg006.rouse-eng2.xml')
bookchap = ''
for l in f:
    m = re.search('subtype="book" n="([0-9]+)">',l)
    if(m):
        curbook = m[1]
    m = re.search('subtype="chapter" n="([0-9]+)">',l)
    if(m):
        curchap = m[1]
        bookchap = curbook + '.' + curchap
        words2.append(curbook + '.' + curchap)
    if(not re.search(citpat,bookchap)):
        continue
    l = re.sub('<[^>]+>',' ',l)
    for foo in l.split():
        words2.append(foo)
outf = open('gpt1','w')
for foo in words2:
    print(foo,file=outf)
outf.close()
```
